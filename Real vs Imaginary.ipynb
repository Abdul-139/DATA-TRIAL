```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the CSV data
df = pd.read_csv("https://raw.githubusercontent.com/Abdul-139/DATA-TRIAL/refs/heads/main/Real%20Vs%20Imaginary.csv")

# Check the first few rows to understand the structure
df.head()
```

```python
# Extract the relevant data
x_values = df['X']
real_values = df['Real']
imaginary_values = df['Imaginary']

# Plotting the NMR spectrum from real and imaginary components

# Real part of the spectrum (typically used for analysis)
plt.figure(figsize=(10, 6))
plt.plot(x_values, real_values, label='Real Part', color='b')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Intensity')
plt.title('NMR Spectrum')
plt.legend()
plt.grid()
plt.show()

# Imaginary part of the spectrum
plt.figure(figsize=(10, 6))
plt.plot(x_values, imaginary_values, label='Imaginary Part', color='r')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Intensity')
plt.title('NMR Spectrum')
plt.legend()
plt.grid()
plt.show()
```
